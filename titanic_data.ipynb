```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Load the datasets
import pandas as pd
train_data = pd.read_csv("C:\\Users\\Rathan\\Downloads\\train.csv")
test_data = pd.read_csv("C:\\Users\\Rathan\\Downloads\\test.csv")
gender_submission = pd.read_csv("C:\\Users\\Rathan\\Downloads\\gender_submission.csv")
print(train_data.head())
```

```python
print("Train data columns:", train_data.columns)
```

```python
test_data = pd.read_csv("C:\\Users\\Rathan\\Downloads\\test.csv")
test_data.head()
```

```python
gender_submission = pd.read_csv("C:\\Users\\Rathan\\Downloads\\gender_submission.csv")
gender_submission.head()
```

```python
print(train_data.isnull().sum())      
```

```python
# Fill missing values
train_data['Age'].fillna(train_data['Age'].median(), inplace=True)
```

```python
train_data['Embarked'].fillna(train_data['Embarked'].mode()[0], inplace=True)
```

```python
test_data['Age'].fillna(test_data['Age'].median(), inplace=True)
test_data['Fare'].fillna(test_data['Fare'].median(), inplace=True)
```

```python
test_data
```

```python
# Convert categorical columns to numeric
train_data['Sex'] = train_data['Sex'].map({'male': 0, 'female': 1})
test_data['Sex'] = test_data['Sex'].map({'male': 0, 'female': 1})
test_data
```

```python
train_data['Embarked'] = train_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
test_data['Embarked'] = test_data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
train_data
```

```python
# Select Features and Target
features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X = train_data[features]
y = train_data['Survived']
print(X.head())
```

```python
print(y.head())
```

```python
#Train a Decision Tree Model
model = DecisionTreeClassifier(random_state=0)
model.fit(X, y)
```

```python
#Predict on test data
X_test = test_data[features]
predictions = model.predict(X_test)
print(predictions[:10])
```

```python
#Prepare submission file
submission = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": predictions
})
print(submission.head(10))
```

```python
merged = submission.merge(gender_submission, on="PassengerId", suffixes=('_model', '_gender'))
print("Comparison with gender-based submission:")
print(merged.head())
```

```python
# Save your final output
submission.to_csv("titanic_submission.csv", index=False)
print("Submission saved as titanic_submission.csv")
```
